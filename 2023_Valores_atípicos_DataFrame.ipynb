```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data= pd.read_excel('Gastos y costos 20-23.xlsx',sheet_name="2023") 
```

```python
#Verificamos información del DataFrame
data.info()
```

```python
#Corroboramos valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

## **PROCEDIMIENTO PARA ELIMINAR VALORES NULOS EN DATAFRAME**

```python
#Reemplazamos valores nulos del dataframe con "bfill"
data1 =data.fillna(method="bfill") 
data1
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores nulos del dataframe con "mean"
#Realizamos una copia del dataframe
data2=data.copy()
data2=data2.fillna(round(data.mean(),1))
data2
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data2.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

##**PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
cuanti2023=data2.drop(['RFC','FECHA','FOLIO','UUID', 'PROVEEDOR','TIPO GASTO','TIPO', 'DESCRIPCION','MP','STATUS','POLIZA'], axis=1)
cuanti2023
```

```python
cuali2023=data2.drop([ 'TC',	'IMPORTE',	'IVA', 'RET ISR','RET ISR2',	'RET IVA', 'TOTAL MX',	'TOTAL SAT' ], axis=1)
cuali2023
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=cuanti2023
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
outliers= cuanti2023[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data3= cuanti2023[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=cuanti2023.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=cuali2023.copy()
data_clean=data_clean.fillna(round(cuali2023.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
# Unimos laos datas con el dataframe
Datos_limpios2023_desv = pd.concat([cuanti2023, cuali2023], axis=1)
Datos_limpios2023_desv
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Datos_limpios2023_desv.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
Datos_limpios2023_desv.to_csv("Datos_limpios2023_desv.csv")
```

##**PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=cuanti2023

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= cuanti2023[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= cuanti2023[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=cuanti2023.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data5_iqr=cuanti2023.copy()
data5_iqr=data5_iqr.fillna(round(cuanti2023.mean(),1))
data5_iqr
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data6_iqr=cuali2023.copy()
data6_iqr=data5_iqr.fillna(round(cuali2023.mean()))
data6_iqr
```

```python
# Unimos laos datas con el dataframe
Datos_limpios2023_iqr = pd.concat([cuanti2023, cuali2023], axis=1)
Datos_limpios2023_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Datos_limpios2023_iqr.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
Datos_limpios2023_iqr.to_csv("Datos_limpios2023_iqr.csv")
```
